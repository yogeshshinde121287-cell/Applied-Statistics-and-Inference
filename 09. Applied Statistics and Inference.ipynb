# Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?

Ans 1 :- Hypothesis testing is a statistical method used to make decisions based on data. We start with two hypotheses
Null Hypothesis (Ho): There is no effect or no difference.
Alternative Hypothesis (Ha): There is an effect or a difference.

After analyzing data, we either reject Ho or fail to reject it. However, mistakes can happen in this decision-making process. These mistakes are called Type I and Type II errors.

Type I Error (False Positive)
Rejecting the null hypothesis when it is actually true.We conclude that something has changed or there is an effect, but in reality, nothing has changed.
Example: A medical test says a healthy person has a disease.

Type II Error (False Negative)
Failing to reject the null hypothesis when it is actually false. We conclude that nothing has changed, but in reality, there is a change.
Example: A medical test says a sick person is healthy.

Type I Error: Causes wrong decisions, unnecessary changes, and financial loss.
Type II Error: Causes missed opportunities, lack of improvement, and competitive disadvantage.

Both errors affect the reliability of decisions. Type I error means acting when we should not, and Type II error means not acting when we should. Understanding and controlling these errors is critical for accurate decision-making in research, business, and healthcare.

# Question 2: What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Ans 2:
The P-value (Probability Value) is a statistical measure used in hypothesis testing to help decide whether to reject the null hypothesis (Ho).
It represents the probability of observing the sample data (or something more extreme) assuming that the null hypothesis is true.
If the null hypothesis were true, the P-value tells us how likely it is to see the results we got.

Hypothesis testing starts with two statements:

Null Hypothesis (Ho): There is no effect or difference.
Alternative Hypothesis (Ha): There is an effect or difference.

We collect data and calculate a test statistic (like Z or t). The P-value is then computed based on this statistic.

Small P-value (e.g., < 0.05):
The observed data is very unlikely under H₀. This gives strong evidence against Ho, so we reject H₀.
Large P-value (e.g., > 0.05):
The observed data is quite likely under H₀. This means weak evidence against Ho, so we fail to reject H₀.

A P-value does NOT tell us the probability that Ho is true.
It only tells us how consistent the data is with Ho.

Usually compare the P-value to a threshold called alpha (α), often set at 0.05 (5%).

If P-value ≤ α: Reject Ho.
If P-value > α: Fail to reject Ho.

Suppose we test if a new medicine works better than the old one:
Ho: Both medicines are equally effective.
After testing, P-value = 0.02.
Since 0.02 < 0.05, we reject Ho and conclude the new medicine is likely better.

Wrong: “P-value = 0.02 means there is a 2% chance Ho is true.”
Correct: “If Ho were true, there is a 2% chance of seeing results this extreme.”

< 0.01 Very strong evidence against Ho, 
0.01 – 0.05 Strong evidence against Ho,
0.05 – 0.10 Weak evidence against Ho, 
> 0.10 No evidence against Ho

The P-value is a key tool in hypothesis testing. It helps us judge whether the observed data provides enough evidence to reject the null hypothesis. A smaller P-value means stronger evidence against Ho, while a larger P-value means we stick with Ho.

# Question 3: Explain the difference between a Z-test and a T-test, including when to use each?

Ans 3 : Z-test and a T-test are both statistical methods used to determine whether there is a significant difference between a sample statistic and a population parameter, but they differ in their assumptions and applications. A Z-test is appropriate when the population standard deviation is known and the sample size is large, typically greater than 30. It relies on the normal distribution, which assumes that with a large sample, the sampling distribution of the mean is approximately normal. This makes the Z-test ideal for large datasets or situations where population parameters are well established. On the other hand, a T-test is used when the population standard deviation is unknown and must be estimated from the sample data. It is most commonly applied to small samples, usually less than 30 observations, where the uncertainty about variability is higher. The T-test uses the t-distribution, which is similar to the normal distribution but has heavier tails to account for this additional uncertainty. As the sample size increases, the t-distribution approaches the normal distribution, making the difference between the two tests negligible for very large samples. In practice, if you have a large sample and know the population variability, you should use a Z-test; if you have a small sample or do not know the population variability, you should use a T-test. Choosing the correct test is critical because using a Z-test for small samples can underestimate variability and lead to incorrect conclusions, while the T-test adjusts for this by incorporating more variability into its calculations.

# Question 4: What is a confidence interval, and how does the margin of error influence its width and interpretation?

Ans 4 : - A confidence interval is a statistical range that provides an estimate of where a population parameter, such as a mean or proportion, is likely to fall based on sample data. Instead of giving a single point estimate, it offers a range of plausible values along with a confidence level, such as 95% or 99%, which indicates how certain we are that the interval contains the true parameter. For example, if we calculate a 95% confidence interval for the average weight of a group as 65 kg to 75 kg, it means that if we repeated the sampling process many times, about 95% of those intervals would contain the true population mean. The width of this interval is influenced by the margin of error, which depends on three main factors: sample size, variability in the data, and the chosen confidence level. A larger margin of error results in a wider confidence interval, signaling greater uncertainty about the estimate, while a smaller margin of error produces a narrower interval, indicating higher precision. The margin of error decreases as sample size increases because larger samples provide more accurate estimates of the population parameter. Conversely, increasing the confidence level from 95% to 99% makes the interval wider because we need to capture more possible values to be more certain. In practical terms, the confidence interval helps decision-makers understand not only the estimate but also the degree of uncertainty associated with it. A narrow interval suggests strong reliability, while a wide interval indicates that more data or better sampling methods may be needed for precise conclusions. Thus, the margin of error plays a critical role in shaping the interpretation of confidence intervals by directly affecting their width and the level of certainty in statistical inference.

# Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Ans 5 : - The purpose of an ANOVA (Analysis of Variance) test is to determine whether there are statistically significant differences between the means of three or more independent groups. While a t-test is limited to comparing two groups, ANOVA extends hypothesis testing to multiple groups simultaneously, reducing the risk of error that would occur if multiple t-tests were performed. ANOVA works by analyzing the variation within each group and the variation between groups. If the variation between groups is much larger than the variation within groups, it suggests that at least one group mean is different from the others. The assumptions of ANOVA are critical for its validity: first, observations must be independent, meaning the data from one group should not influence another; second, the data within each group should be approximately normally distributed; and third, the variances across groups should be equal, known as homogeneity of variance. These assumptions ensure that the F-statistic, which ANOVA uses to test differences, is accurate. ANOVA does not tell us which specific groups differ; it only indicates that a difference exists, and post-hoc tests like Tukey’s test are used to identify where the differences lie. In summary, ANOVA is a powerful extension of hypothesis testing that allows researchers to compare multiple groups in a single test, making it essential in experiments involving more than two conditions or treatments.

# Question 6: Python program to perform a one-sample Z-test and interpret the result for a given dataset.

```python
import numpy as np
from scipy import stats

sample = np.array([50, 52, 49, 51, 50, 53, 48])
mu = 50  # population mean
sigma = 2  # population standard deviation
n = len(sample)

z = (np.mean(sample) - mu) / (sigma / np.sqrt(n))

p_value = 2 * (1 - stats.norm.cdf(abs(z)))

print(f"Z-statistic: {z:.4f}, P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: Sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

# Question 8: Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.


```python
import numpy as np
import plotly.express as px

samples_means = []
for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    samples_means.append(np.mean(sample))

fig = px.histogram(samples_means, nbins=30, title='Sample Means Approaching Normality (CLT)')
fig.show()
```

# Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.


```python
import numpy as np
import plotly.express as px

sample = np.random.normal(50, 10, 100)
mean = np.mean(sample)
std = np.std(sample, ddof=1)
confidence = 0.95
z = stats.norm.ppf(1 - (1-confidence)/2)
margin_error = z * (std / np.sqrt(len(sample)))
ci_lower, ci_upper = mean - margin_error, mean + margin_error

print(f"Mean: {mean:.2f}, 95% CI: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

# Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.


```python
from scipy.stats import chisquare

observed = [50, 30, 20]
expected = [40, 40, 20]
chi2, p = chisquare(observed, f_exp=expected)
print(f"Chi-square statistic: {chi2:.4f}, P-value: {p:.4f}")
if p < 0.05:
    print("Reject H0: Observed distribution differs significantly from expected.")
else:
    print("Fail to reject H0: No significant difference.")
```
